```python
path = "/content/drive/MyDrive/Colab Notebooks/skema/ASKEM Grounding Annotations"
files = [os.path.join(path, file) for file in os.listdir(path)]
files
```

```python
label_counts = {}
role_based_counts = {}
candidate_counts = {}
unique_text_counts = {}
tot_text_counts = {}
for file in files:
  df = pd.read_excel(file)
  df['Text'] = df['Text'].fillna("NA")
  df['Role'] = df['Role'].fillna("NA")
  df['Context'] = df['Context'].fillna("NA")
  df['Annotation'] = df['Annotation'].fillna("")
  # df['Identified_type'] = df['Identified_type'].fillna("")
  # df['Comments'] = df['Comments'].fillna("")
  prev_text = ""
  prev_role = ""
  prev_context = ""
  for index, this_row in df.iterrows():
    this_row_list = this_row.values.tolist()
    if this_row['Text'] != "NA" and this_row['Role'] != "NA" and this_row['Context'] != "NA":
      prev_text = this_row['Text']
      prev_role = this_row['Role']
      prev_context = this_row['Context']
    else:
      df.at[index, 'Text'] = prev_text
      df.at[index, 'Role'] = prev_role
      df.at[index, 'Context'] = prev_context
  print("*********For %s, the label counts are "%(file))
  gp = df.groupby(["Annotation"]).count().reset_index()
  gp = gp.rename(columns={"Annotation":"Annotation Label", "Text":"# Count"})
  gp.at[0,"Annotation Label"] = "None"
  label_counts[file] = gp[["Annotation Label", "# Count"]]

  gp = df.groupby(["Role", "Annotation"]).count().reset_index()
  gp = gp.rename(columns={"Annotation":"Annotation Label", "Text":"# Count"})
  role_based_counts[file] = gp[["Role","Annotation Label", "# Count"]]

  candidate_counts[file] = len(df["Candidate ID"].unique())
  unique_text_counts[file] = len(df["Text"].unique())
  tot_text_counts[file] = len(df)
```
